```python
import pandas as pd
import math
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

```python
df_merge = pd.read_csv('full_merge.csv')
df_merge = df_merge.dropna()
df_merge.head()
```

```python
df_merge['sentiment'] = df_merge['Reviews'].astype(str).apply(lambda x: sid.polarity_scores(x)['compound'])
```

```python
df_merge.to_csv('vader_sentiment.csv',index=False)
```

### Scaling Sentiment rating

```python
sentiment_df = pd.read_csv('vader_sentiment.csv')
```

```python
sentiment_df['sentiment'] = sentiment_df['sentiment'].apply(lambda x: math.ceil(x*10))
```

```python
sentiment_df.loc[(sentiment_df['sentiment'] == -10) | (sentiment_df['sentiment'] == -9) , 'updated_sentiment'] = 1
sentiment_df.loc[(sentiment_df['sentiment'] == -8) | (sentiment_df['sentiment'] == -7) , 'updated_sentiment'] = 2
sentiment_df.loc[(sentiment_df['sentiment'] == -6) | (sentiment_df['sentiment'] == -5) , 'updated_sentiment'] = 3
sentiment_df.loc[(sentiment_df['sentiment'] == -4) | (sentiment_df['sentiment'] == -3) , 'updated_sentiment'] = 4
sentiment_df.loc[(sentiment_df['sentiment'] == -2) | (sentiment_df['sentiment'] == -1) | (sentiment_df['sentiment'] == 0) , 'updated_sentiment'] = 5
sentiment_df.loc[(sentiment_df['sentiment'] == 0) | (sentiment_df['sentiment'] == 1) | (sentiment_df['sentiment'] == 2) , 'updated_sentiment'] = 6
sentiment_df.loc[(sentiment_df['sentiment'] == 3) | (sentiment_df['sentiment'] == 4) , 'updated_sentiment'] = 7
sentiment_df.loc[(sentiment_df['sentiment'] == 5) | (sentiment_df['sentiment'] == 6) , 'updated_sentiment'] = 8
sentiment_df.loc[(sentiment_df['sentiment'] == 7) | (sentiment_df['sentiment'] == 8) , 'updated_sentiment'] = 9
sentiment_df.loc[(sentiment_df['sentiment'] == 9) | (sentiment_df['sentiment'] == 10) , 'updated_sentiment'] = 10
```

```python
sentiment_df = sentiment_df.rename(columns={'updated_sentiment':'Sentiment'})
sentiment_df = sentiment_df.drop(columns=['sentiment'])
```
